# Data Wrangling

## Handle Missing Values

```python
#Mencari kolom yang mempunyai missing value dalam bentuk persen
missing = df.isnull().sum()*100/df.shape[0]
missing[missing>0].sort_values()
```

Setelah melihat persentase missing value pada tiap kolom maka akan dilakukan perlakuan sederhana seperti berikut:
- Membuang kolom dengam missing value >50%
- Membuang baris dimana terdapat kolom yang kosong <10%
- Mengimpute kolom numerik dengan 0

```python
#Mencari kolom dan baris yang perlu dihapus
drop_cols = missing[missing>50]
drop_row = missing[missing<10]
print('Dataframe shape before: '+str(df.shape))
#Membuang kolom
df = df.drop(drop_cols.index.to_list(),axis=1)
#Membuang baris
df = df.dropna(subset=drop_row.index.to_list())

#Mencari kolom numerik pada dataframe baru
numerik_cols = []
for col in df.columns.to_list():
    if (df[col].dtype == 'int64') or (df[col].dtype == 'float'):
        numerik_cols.append(col)
        
#Impute 0
df = df.fillna(0)
print('Dataframe shape after: '+str(df.shape))
print('Missing values pada dataframe')
missing = df.isnull().sum()*100/df.shape[0]
missing[missing>0].sort_values()
```

## Feature Transformation

Setelah melihat beberapa fitur akhir dataset dapat dilihat banyak nilai yang menunjukan "x+y". Hal ini bisa ditransformasi menjadi bentuk numerik yang dapat lebih merepresentasikan arti fitur tersebut.

```python
#Membuat fungsi transformasi ke bentuk numerik
def transform(x):
    x = str(x)
    list_x = x.split('+')
    count = 0
    for i in list_x:
        count += int(i)
    return count
```

```python
#Mengaplikasikan fungsi transformasi ke fitur-fitur akhir
columns = df.columns.to_list()
index = columns.index('ls')
for i in range(index,len(columns)):
    df[columns[i]] = df[columns[i]].apply(transform)
df.head()
```

Lalu selanjutnya dapat diubah fitur 'joined' menjadi tipe data waktu. Hal ini karena fitur tersebut berarti waktu pemain tersebut bergabung ke dalam klub dan akan lebih tepat jika direpresentasikan dalam bentuk waktu

```python
from datetime import datetime

#Mengubah tipe data joined menjadi tipe waktu
df['joined'] = pd.to_datetime(df['joined'])
df['joined'].head()

#Mencari tahun joined
df['joined_years'] = df['joined'].apply(lambda x: x.year)
df['joined_years'].head()
```

# Unsupervised Learning

Dengan Machine Learning akan dilihat bagaimana model dengan metode K Means Clustering mengelompokkan posisi pemain berdasarkan kemampuan mereka. ML ini hanya akan dilakukan untuk 6 posisi terbanyak pada dataset. Selain itu tidak dilakukan train dan test karena hanya akan dilakukan ML sederhana dan sulit mengukur akurasi clustering untuk label >2.

```python
pos = df['team_position'].value_counts().sort_values(ascending=False)
df_top_pos = pd.DataFrame()
for p in pos.keys()[:6]:
    df_top_pos = pd.concat([df[df['team_position']==p],df_top_pos])
```

```python
cols = df.columns.to_list()
start = cols.index('pace')
end = cols.index('goalkeeping_reflexes')
cols = ['international_reputation','weak_foot','skill_moves','overall','potential']+cols[start:end+1]
X = df_top_pos[cols]
X = X.to_numpy()
```

```python
# Elbow Method
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
  
# Visualize 
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

Sebenarnya tidak perlu dilakukan Elbow Method karena akan dilakukan validasi prediksi label model dengan posisi pemain yang asli. Karena pada kasus ini posisi pemain hanya ada 6 maka akan ada 6 buah cluster untuk modelling. Namun jika menggunakan elbow method yang sebenarnya harusnya digunakan n = 4

```python
model = KMeans(n_clusters=6)
model.fit(X)
```

```python
# aplikasikan model ke dataset yang kita punya tadi
labels = model.predict(X)
df_top_pos['labels'] = labels
df_top_pos.head()
```

```python
evaluasi = pd.DataFrame({'labels': labels, 'pos': df_top_pos['team_position']})
ct = pd.crosstab(evaluasi['labels'], evaluasi['pos'])
ct
```

Model nampaknya tidak dapat memprediksi posisi asli pemain berdasarkan score dari aspek-aspek kemampuan pemain tersebut. Hal ini dapat dilihat lebih jelas dibawah.

```python
row=0
fig , axs = plt.subplots(3,2,figsize=(17,12))
for i in range(3):
    for j in range(2):
        sns.barplot(ax=axs[i,j],x=ct.columns.to_list() ,y=ct.iloc[row].values,palette='RdBu')
        axs[i,j].set_title('Category '+str(row))
        row+=1   
```

Dapat dilihat pada grafik di atas bahwa nampaknya pemain dengan posisi SUB terlalu banyak pada dataset yang digunakan dalam modeling sehingga model kebanyakan memprediksi posisi SUB pada tiap labels.
