# 3. Modelling
disini akan dibuat model machine learning yang memprediksi nilai pentingnya physical touch sama calon pacar berdasarkan 5 variable lainnnya

```python

# pada modelling ini, features yang digunakan adalah :
features = ['Cewek atau cowok nih?','Seberapa penting physical touch sama calon pacar untuk kamu?','Seberapa penting word of affirmation dari calon pacar untuk kamu?','Seberapa penting dapet kado dari calon pacar untuk kamu?','Seberapa penting bantuan dari calon pacar untuk kamu?']
X = data[features]
# karena data cewek atau cowok berupa object, maka dibuat integer 0,1 dulu
X['Cewek atau cowok nih?']=X['Cewek atau cowok nih?'].replace('Cowok',1)
X['Cewek atau cowok nih?']=X['Cewek atau cowok nih?'].replace('Cewek',0)

# dengan target yang diprediksi adalah :
y = data['Seberapa penting quality time bareng calon pacar untuk kamu?']

# agar prediksi bisa diukur dengan baik (tidak overfit), maka dibuat 2 bagian datanya
# dan di set random_state=1 untuk ketika di run lagi tidak mengubah nilai
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1) 
# variabel dengan awalan train_ itu untuk training model
# variabel dengan awalan val_ itu untuk evaluasi model


# menggunakan max_leaf_nodes yang menghasilkan nilai terkecil, yaitu 4
model = RandomForestRegressor(max_leaf_nodes=4, random_state=0) 

# fitting model
model.fit(train_X, train_y)
preds_val = model.predict(val_X)

# melihat seberapa akurat model berdasarkan persentase keakuratan model
print(f"Persentase keakurasian model adalah : {100 - mean_absolute_error(val_y, preds_val)*100}%")
```
