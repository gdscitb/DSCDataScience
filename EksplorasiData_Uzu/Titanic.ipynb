# mengambil title dari kolom name dan dimasukkan kedalam kolom title



```python
df['title'] = df['Name'].str.extract(' ([A-Za-z]+)\.')
df_test['title'] = df_test['Name'].str.extract(' ([A-Za-z]+)\.')

display(df.head())
display(df_test.head())
```

kita akan mengganti kolom title dengan numerik

Mr --> 0

Miss --> 1

Mrs --> 2

Master, dll --> 3

```python
# mengganti title tersebut menjadi numerik
replace = {'Mr':0, 'Miss':1, 'Mrs':2, 'Master':3, 'Dr':3, 'Rev':3, 'Col':3, 'Major':3, 'Mlle':3, 'Countess':3, 'Jonkheer':3,
          'Capt':3, 'Don':3, 'Ms':3, 'Lady':3, 'Mme':3, 'Sir':3, 'Dona':3}

df['title'] = df['title'].replace(replace)
df_test['title'] = df_test['title'].replace(replace)

df.head(10)
```

dapat dilihat bahwa kolom age dari kedua dataframe terdapat missing values

sehingga kita harus menangani missing values tersebut

```python
import seaborn as sns

plt.subplots(figsize=(10,6), dpi=100)
sns.boxplot(y=df['Age'], showmeans=True,
            meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                       "markersize":"10"})
plt.show()
```

Karena kolom age terdapat outlier maka saya akan menangani missing value dengan menggunakan median

```python
# mengisi masing-masing missing value dengan median berdasarkan masing-masing title nya
df['Age'].fillna(df.groupby('title')['Age'].transform('median'), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# Hal yang sama juga kita lakukan untuk menangani missing values di data test
df_test['Age'].fillna(df_test.groupby('title')['Age'].transform('median'), inplace=True)
df_test.head(10)
```


Melakukan Binning/Diskritisasi pada kolom age

umur <= 16 : 0

16 < umur <= 26 : 1

26 < Umur <= 36 : 2

36 < umur <= 56 : 3

umur > 56 : 4

```python
df.loc[df.Age <= 16, 'Age'] = 0
df.loc[(df.Age > 16) & (df.Age <= 26), 'Age'] = 1
df.loc[(df.Age > 26) & (df.Age <= 36), 'Age'] = 2
df.loc[(df.Age > 36) & (df.Age <= 56), 'Age'] = 3
df.loc[df.Age > 56, 'Age'] = 4
```

```python
df.head(10)
```

```python
df_test.loc[df_test.Age <= 16, 'Age'] = 0
df_test.loc[(df_test.Age > 16) & (df_test.Age <= 26), 'Age'] = 1
df_test.loc[(df_test.Age > 26) & (df_test.Age <= 36), 'Age'] = 2
df_test.loc[(df_test.Age > 36) & (df_test.Age <= 56), 'Age'] = 3
df_test.loc[df_test.Age > 56, 'Age'] = 4

df_test.head(10)
```

Kolom Embarked

C : 1

Q : 2

```python
df['Embarked'] = df['Embarked'].replace({'S':0, 'C':1, 'Q':2})
df_test['Embarked'] = df_test['Embarked'].replace({'S':0, 'C':1, 'Q':2})
```

```python
# karena pada kolom embarked masih ada missing value maka harus kita tangani
df['Embarked'].fillna(df.Embarked.median(), inplace=True)
```

```python
#Menggabungkan kolom SibSp dan Parch menjadi kolom Family

df['Family'] = df['SibSp'] + df['Parch']
df_test['Family'] = df['SibSp'] + df['Parch']
```

```python
#menangani kolom Fare

df['Fare'].fillna(df.groupby('Pclass')['Fare'].transform('median'), inplace=True)
df_test['Fare'].fillna(df_test.groupby('Pclass')['Fare'].transform('median'), inplace=True)
```

Binning pada kolom Fare

fare $\le$ 17 : 0

17 < fare $\le$ 30 : 1

30 < fare $\le$ 100 : 2

fare > 100 : 3

```python
df.loc[df['Fare'] <= 17, 'Fare'] = 0
df.loc[(df['Fare'] > 17) & (df['Fare'] <= 30), 'Fare'] = 1
df.loc[(df['Fare'] > 30) & (df['Fare'] <= 100), 'Fare'] = 2
df.loc[df['Fare'] > 100, 'Fare'] = 3
display(df.head(10))

df_test.loc[df_test['Fare'] <= 17, 'Fare'] = 0
df_test.loc[(df_test['Fare'] > 17) & (df_test['Fare'] <= 30), 'Fare'] = 1
df_test.loc[(df_test['Fare'] > 30) & (df_test['Fare'] <= 100), 'Fare'] = 2
df_test.loc[df_test['Fare'] > 100, 'Fare'] = 3
display(df_test.head(10))
```

# 4. Membuat Model

```python
# membuat kolom feature dan kolom label
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
# menggunakan model KNN
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score

knn = KNeighborsClassifier()
params = {
    'n_neighbors' : np.arange(1,50),
    'weights' : ['uniform', 'distance']
    }

knn_gscv = GridSearchCV(knn, param_grid=params , cv=5, scoring='accuracy') 
knn_score = cross_val_score(knn, X, y , cv=5, scoring='accuracy')
print(knn_score.mean())
```

```python
# menggunkan model Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
params = {
    'criterion' : ["gini", "entropy"],
    'splitter' : ["best", "random"]
}

dtc_gscv = GridSearchCV(dtc, param_grid = params, cv=5, scoring='accuracy')
dtc_score = cross_val_score(dtc_gscv, X,y, cv=5 , scoring = 'accuracy')
print(dtc_score.mean())
```

```python
# menggunakan model Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
params = {
    'criterion' : ["gini", "entropy"]
}

rfc_gscv = GridSearchCV(rfc, param_grid = params, cv=5, scoring='accuracy', verbose=True)
score_rfc = cross_val_score(rfc_gscv, X, y, cv=5, scoring='accuracy')
print(score_rfc.mean())
```

```python
# menggunakan model SVM
from sklearn.svm import SVC

svc = SVC()
params = {
    'gamma' : ['scale', 'auto'],
    'C' : [1000,10000,100000],
}

svc_gscv = GridSearchCV(svc, param_grid = params, cv=5, scoring='accuracy', verbose=True)
svc_score = cross_val_score(svc_gscv, X,y, cv=5, scoring='accuracy')
print(svc_score.mean())
```

```python
# menggunakan model naive bayes classifier
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb_score = cross_val_score(nb, X, y, cv=5, scoring='accuracy', verbose=True)
print(nb_score.mean())
```

```python
# menggunakan model Gaussian Process Classifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF

gp = GaussianProcessClassifier(1.0 * RBF(1.0))
gp_score = cross_val_score(gp, X, y , scoring='accuracy', verbose=True)
print(gp_score.mean())
```

# 5. Testing

```python
# karena accuracy nya lebih bagus Gaussian Process Classifier maka akan digunakan model tersebut
gp.fit(X,y)

test_data = df_test.drop('PassengerId', axis=1).copy()
predict = gp.predict(test_data)
```

```python
gp_score_test = cross_val_score(gp, test_data, predict, scoring='accuracy')
print(gp_score_test.mean())
```

```python
submission = pd.DataFrame({
    'PassengerId' : df_test['PassengerId'],
    'Survived' : predict
})
# menyimpan hasil prediksi kedalam csv
submission.to_csv('submission.csv', index=False)
```
