```
import pandas as pd
import seaborn as sns
import numpy as np
```

```
df = pd.read_csv('/content/sample_data/video_games.csv')
```

```
df.head()
```

#Data Preparation

```
df = df.iloc[:, :16]
```

```
df.columns = ['title','handheld','nplayer','multiplatform','online','genre','license','publisher','sequel','revscore','sales','price','console','rating','release','year']
```

```
df.head()
```

#Data Analysis

###Highest Sales

```
df[df.sales==df.sales.max()]
```

###Highest price

```
df[df.price==df.price.max()]
```

###Highest review score

```
df[df.revscore==df.revscore.max()]
```

###Smallest sales

```
df[df.sales==df.sales.min()]
```

###Smallest price

```
df[df.price==df.price.min()]
```

###Smallest review score

```
df[df.revscore==df.revscore.min()]
```

###Average sales

```
mean_sales=df['sales'].mean()
print('Average sales:',mean_sales)
```

###Average Price

```
mean_price=df['price'].mean()
print('Average price:',mean_price)
```

###Average review score

```
mean_rs=df['revscore'].mean()
print('Average review score:',mean_rs)
```

###Sales standard deviation

```
std_sales=df['sales'].std()
print('The standard deviation of video games sales:',std_sales)
```

```
std_price=df['price'].std()
print('The standard deviation of video games price:',std_price)
```

```
std_rs=df['revscore'].std()
print('The standard deviation of video games review scores:',std_rs)
```

###Top 5 Genres

```
df['genre'].value_counts().nlargest(5)
```

#Data Visualization

```
sns.scatterplot(x='revscore', y='sales',hue='publisher', data=df)
```

```
sns.countplot(df['genre'])
```

#Data Correlation

```
df.corr(method='pearson')
```

```
df.corr(method='kendall')
```



*   Data correlation with pearson and kendall shows that the correlation between review score and sales are always a low positive correlation.
*   That means there is a correlation, but not a significant one.
*   On the other hand, correlation between review score and year is a low negative correlation. 
*   That means it has negative to none correlation.
