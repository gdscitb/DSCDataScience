## Analisis Dasar

```python
df.sort_values(['Country', 'Brand'])
```

```python
varieties = df.groupby(["Country"]).count()
```

```python
del varieties['Style']
del varieties['Brand']
```

```python
rank = df.groupby(["Country"]).mean()
```

```python
result = pd.merge(varieties, rank, how='right', on=['Country'])
```

```python
result.sort_values(['Stars'], ascending = False)
```

```python
result.sort_values(['Variety'], ascending = False)
```

```python
stylerank = df.groupby(["Style"]).mean()
stylemany = df.groupby(["Style"]).count()
styleresult = pd.merge(stylerank, stylemany, how='right', on=['Style'])
del styleresult['Country']
del styleresult['Brand']
del styleresult['Stars_y']
styleresult.sort_values(['Variety'], ascending = False)
```

# Conclusion

### As we can see from the tables (which have been sorted) above, we can conclude these things :
- The most common style of ramen throughout the world is pack, and it has the best ratings too compared to the other styles (exclusing Box and Bar because it has too few varieties)
- Countries that have invented the most varieties of ramen are Japan, USA, South Korea, Indonesia, Malaysia, China. We can conclude that ramen are more popular in Asian countries hence the result why there are more varieties of ramen there
- Countries that have the best ramen quality (deduced from the star ratings) are South american countries. However, there are too few varieties so it could be an anomaly. Excluding those countries, we get that south-east asian like Indonesia Malaysia and Singapore (followed by japan) are the countries that provides the best ramen quality.
