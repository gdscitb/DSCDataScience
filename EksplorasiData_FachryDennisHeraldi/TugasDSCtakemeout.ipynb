# Penggunaan Library

```
import pandas as pd
from sklearn.neighbors import NearestNeighbors

import pickle
```

# Mulai Mengimport Data

```
df = pd.read_csv('/content/takemeout.csv')
```

```
df.head()
```

# Data Preprocessing

Data cleaning

```
df = df.iloc[:, 1:]
#Menghapus kolom Timestamp karena tidak diperlukan dalam pengolahan data
```

```
df.head()
```

```
df.columns = ['name', 'gender','qt', 'touch', 'words', 'gift','service']
#Mengubah nama kolom menjadi kata yang mudah dimengerti agar memudahkan dalam pengolahan data
```

```
df.head()
```

```
df['gender']=df['gender'].apply(lambda x: 0 if x== 'Cowok' else 1)
#Gender cewek direpresentasikan dengan 1 sedangkan cowok dengan 0
```

```
df.head()
```

```
df.describe()
```

Berdasarkan data diatas maka didapatkan bahwa partisipan mementingkan quality time dibandingkan dengan aspek yang lain. Dilihat dari rata-rata tertinggi terdapat pada qualiy time yaitu 4.099. Sedangkan gift adalah aspek yang tidak begitu dipentingkan karena memiliki nilai rata-rata terendah, yaitu hanya 2.73

```
m_df = df[df['gender'] == 0]
f_df = df[df['gender'] == 1]
```

```
m_x = m_df.drop(['name','gender'],axis=1)
f_x = f_df.drop(['name','gender'],axis=1)
```

# Model Development

Akan dicoba permodelan data dengan menyocokkan nilai aspek-aspek terdekat/mirip dengan satu partisipan dengan partisipan lainnya

```
m_nbrs = NearestNeighbors(n_neighbors=5).fit(m_x)
f_nbrs = NearestNeighbors(n_neighbors=5).fit(f_x)
```

```
X = df.drop(['name','gender'], axis=1)
```

```
test = X[df['name'] == 'f*****'].values
```

```
_, idx = f_nbrs.kneighbors(test)
```

```
idx
```

```
f_df.iloc[idx[0]]['name']
```

```
pickle.dump(m_nbrs, open('/content/m_model.pickle', 'wb'))
pickle.dump(f_nbrs, open('/content/f_model.pickle', 'wb'))
```
