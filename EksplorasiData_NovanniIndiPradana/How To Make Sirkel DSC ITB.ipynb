```python
df.columns= ['Nama', 'Kelas', 'Tipe', 'Waktu' , 'IPK', 'Tugas', 'Materi']
```

```python
df.head()
```

```python
df['Kelas'] = df['Kelas'].apply(lambda x: 0 if x == 'D4-KS' else 1  )
```

```python
ks_df = df[df['Kelas'] == 0]
st_df = df[df['Kelas'] == 1]
```

```python
ks_X = ks_df.drop(['Nama', 'Kelas'], axis = 1)
st_X = st_df.drop(['Nama', 'Kelas'], axis = 1)
```

# MODEL DEVELOPMENT

Disini saya membuat pengelompokan beberapa orang berdasarkan kesamaan mereka melalui pendekatan kebiasaan dengan tujuan mereka dapat menemukan orang yang sefrekuensi dengan diri mereka masing-masing agar dapat berteman dan bereklompok.

```python
ks_nbrs = NearestNeighbors(n_neighbors=5).fit(ks_X)
st_nbrs = NearestNeighbors(n_neighbors=5).fit(st_X)
```

```python
X = df.drop(['Nama','Kelas'], axis = 1 )
```

```python
test = X[df['Nama'] == 'Wa ode raksni'].values
```

```python
_, idx= ks_nbrs.kneighbors(test)
```

```python
import matplotlib.pyplot as plt


x_ax = ["Tipe", "Waktu", "IPK", "Tugas", "Materi"]
y_ax = [df.Tipe.mean(), df.Waktu.mean(), df.IPK.mean(), df.Tugas.mean(), df.Materi.mean()]
y_ks = [ks_df.Tipe.mean(), ks_df.Waktu.mean(), ks_df.IPK.mean(), ks_df.Tugas.mean(), ks_df.Materi.mean()]
y_st = [st_df.Tipe.mean(), st_df.Waktu.mean(), st_df.IPK.mean(), st_df.Tugas.mean(), st_df.Materi.mean()]


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Rata-rata preferensi KS dan ST")

plt.plot(x_ax, y_ax, label="Umum")
plt.plot(x_ax, y_ks, label="KS")
plt.plot(x_ax, y_st, label="ST")

ax.legend()
```
