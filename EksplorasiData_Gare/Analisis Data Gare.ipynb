# Data preparation and preprocessing

Hal yang pertama perlu kita lakukan ialah mengimport libraries untuk menggunakan fitur fitur yang ada matplotlib digunakaan untuk grafik, seaborn digunakan untuk mempercantik grafik tersebut. pandas digunakan untuk membaca data.

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

Membaca data menggunakan library pandas

```python
data_clean = pd.read_csv("haberman.csv", names = ["Age", "Operation", "Axil_nodes", "Surv_status"])
```

Gambaran kecil mengenai data yang akan kita olah

```python
data_clean.head()
```

Age adalah umur pasien\
Operation adalah tahun terjadinya operasi dalam 1900an
Nodus axil adalah (kelenjar lymph) yang terdeteksi 

Survival status adalah tingkat kebertahanan hidup yang memiliki 2 index\
index 1 merupakan pasien yang bisa bertahan selama 5 tahun atau lebih\
index 2 melambangkan pasien yang mati dalam waktu kurang dari 5 tahun

```python
data_clean.describe()
```

Terlihat bahwa rata - rata usia yang terdapat pada data ini ialah berkisar 52 Tahun

```python
data_clean.info()
```

# Eksplorasi Data

```python
sns.color_palette("bright", as_cmap = False)
sns.set_palette("bright")
```

```python
sns.lmplot(x = 'Age', y = 'Operation', data = data_clean, fit_reg = False, hue = 'Surv_status')
```

```python
sns.displot(data=data_clean, x="Age", col="Surv_status", hue = 'Surv_status', kde=True)
```

Perhatikan grafik persebaran pertama.\
Bisa kita lihat dengan jelas bahwa pasien pada usia sekitar 50 dan 60 memilki tingkat kebertahanan hidup lebih besar

Pada grafik pertama pun kita dapat menyimpulkan bahwa pasien yang melakukan operasi dari tahun 60 hingga 62 memiliki tingkat bertahan hidup lebih baik dibandingkan tahun tahun lainnya

Agar lebih jelas perhatikan persebaran dibawah ini

```python
sns.boxplot(x = 'Surv_status', y = 'Age', data = data_clean)
```

Pada grafik box ini sangat jelas bahwa pasien pada usia 50 hingga 60 memiliki lebar yang lebih besar dibandingkan box sebelahnya akibatnya pada rentang usia ini banyak orang yang memiliki tinkgat bertahan hidup lebih baik

```python
sns.violinplot(x = 'Surv_status', y = 'Operation', data = data_clean)
```

Pada distribusi ini ditunjukan lebih jelas bahwa pada tahun operasi 58 hingga 66 memiliki tingkat bertahan hidup yg lebih baik

```python
sns.pairplot(data_clean, hue = 'Surv_status', height = 3)
```
