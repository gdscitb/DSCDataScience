```
data.columns = ['Date','Avg_Temperature(C)','Lowest_Temperature(C)','Highest_Temperature(C)', 'Precitipation(mm)', 'weekend','Beer_Consumption']
```

```
data['Date'] = pandas.to_datetime(data['Date'])
data['Avg_Temperature(C)'] = data['Avg_Temperature(C)'].str.replace(",",".")
data['Avg_Temperature(C)'] = pandas.to_numeric(data['Avg_Temperature(C)'])
data['Highest_Temperature(C)'] = data['Highest_Temperature(C)'].str.replace(",",".")
data['Highest_Temperature(C)'] = pandas.to_numeric(data['Highest_Temperature(C)'])
data['Lowest_Temperature(C)'] = data['Lowest_Temperature(C)'].str.replace(",",".")
data['Lowest_Temperature(C)'] = pandas.to_numeric(data['Lowest_Temperature(C)'])
```

```
bc_data = data[data['Beer_Consumption'] > 0]
```

```
bc_data.set_index('Date',inplace=True)
bc_data['Beer_Consumption'].plot()
```

```
plt.title('Beer Consumption in a year')
bc_data['Beer_Consumption'].rolling(30).mean().plot()
```

```
temp_data = data[data["Avg_Temperature(C)"] > 0 ]
temp_data.set_index('Date',inplace=True)
plt.title('Avg Temperature of beers consumed in a year')
temp_data['Avg_Temperature(C)'].rolling(25).mean().plot()
```

```
bc_data = data[data['weekend'] == 1]
bc_data.set_index('Date',inplace=True)
plt.title('Beer Consumption during weekends')
bc_data['Beer_Consumption'].rolling(30).mean().plot()
```

```
bc_data = data[data['weekend'] == 0]
bc_data.set_index('Date',inplace=True)
plt.title('Beer Consumption during weekdays')
bc_data['Beer_Consumption'].rolling(30).mean().plot()
```
