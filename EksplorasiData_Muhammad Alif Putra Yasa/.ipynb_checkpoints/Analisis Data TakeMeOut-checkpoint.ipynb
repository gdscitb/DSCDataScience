## Data Cleaning

Langkah pertama yang dilakukan adalah meng-*import* *library* dan membaca data dari file .csv

```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as spsts
import collections as col

# Membaca Data
data = pd.read_csv('takemeout.csv')
```

```python
# Melihat data
data.head()
```

Data yang tidak digunakan di analisis, seperti *Timestamp* dan Nama (Semua nama disensor, sehingga tidak berguna), dihapus atau disingkat.

```python
# Membenahi nama kolom
del data['Timestamp']
column_names = ['Nama', 'Gender', 'Quality Time', 'Physical Touch', 'Affirmation', 'Kado', 'Bantuan']
data.columns = column_names

# Menyingkat Nama
for i in data['Nama']:
    if len(i) > 2:
        data['Nama'] = data['Nama'].replace( i, i[:1] + str(len(i) - 2) + '*')
        
# Memisahkan data cowok dan cewek
cowok_data = data.loc[data['Gender'] == 'Cowok']
cewek_data = data.loc[data['Gender'] == 'Cewek']
```

```python
data.head()
```

## Data Analysis

Data yang diperoleh berbentuk *ordinal data*

### Statistik Dasar

```python
# Statistik dasar data keseluruhan
sts_dasar = {
    "Max"    : [ np.max(data[kolom]) for kolom in column_names[2:] ],
    "Min"    : [ np.min(data[kolom]) for kolom in column_names[2:] ],
    "Mode"   : [ int(spsts.mode(data[kolom])[0]) for kolom in column_names[2:] ],
    "Median" : [ int(np.median(data[kolom])) for kolom in column_names[2:] ]
}

tabel_sts       = pd.DataFrame(sts_dasar)
tabel_sts.style.set_caption("Hello World")
tabel_sts.index = column_names[2:]
tabel_sts
```

## Visualisasi Data

```python
# Mengatur ukuran
plt.rcParams['figure.figsize'] = [2 *10 *1.6 /2.54, 6 *10 /2.54]

#   Untuk membuat judul umum bar chart dan pie chart, 
# langkah awal yang dilakukan adalah membuat 5 Grafik

fig, main_axes = plt.subplots(nrows=5, ncols=1, sharey=True)
plt.subplots_adjust(hspace=0.3)

for row, main_ax in enumerate(main_axes, start=1):
    # Memberi nama grafik
    main_ax.set_title(column_names[2:][row-1], fontsize=15)
    # Membuat grafik tidak terlihat
    main_ax.axis('off')

#   Setelah membuat 5 grafik yang tidak terlihat tetapi judulnya
# tetap terlihat, langkah selanjutnya adalah mengisinya dengan
# bar chart dan pie chart
    
# Membuat 10 plot
for i in range(10):
    # Membuat subplot dengan 5 row dan
    # 2 kolom. i + 1 adalah index plot
    sub_ax = fig.add_subplot(5,2,i+1)
    
    # Data dihitung dan diurut
    hitungData = col.Counter(data[column_names[2:][i//2]])
    urutHitData = sorted(hitungData.items())
    
    # Untuk plot di kiri, bar chart
    if i % 2 == 0:
        sub_ax.bar([i[0] for i in urutHitData],
                [i[1] for i in urutHitData],
                color = 'skyblue')
        plt.setp(sub_ax,
                 xlabel='Tingkat Kesukaan',
                 ylabel = 'Frekuensi')
    
    # Untuk plot di kanan, pie chart
    else:
        sub_ax.pie([i[1] for i in urutHitData],
               labels=[i[0] for i in urutHitData],
               autopct='%1.1f%%')

# Supaya ukurannya pas dengan aturan ukuran
plt.tight_layout()

plt.show()
```

## Insights

Dengan menggunakan [Spearman rank correlation](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient), kita dapatkan bahwa:

```python
# Pasangan Korelasi
aspect = []

# Koefisien Korelasi
coeff = []

# Interpretasi
interp = []

for i in range(4):
    for k in range(i+1, 5):
        aspect += [column_names[2:][i] + ' & ' + column_names[2:][k]]
        res = spsts.spearmanr(data[column_names[2:][i]], data[column_names[2:][k]])[0]
        
        # Koefisiennya 3 angka belakang koma aja
        coeff += [round(res,3)]
        
        # Hasil yang memungkinkan
        if res > 0.9:
            verdict = 'Very high'
        elif res > 0.7 and res <= 0.9:
            verdict = 'High'
        elif res > 0.5 and res <= 0.7:
            verdict = 'Moderate'
        elif res > 0.3 and res <= 0.5:
            verdict = 'Low'
        else:
            verdict = 'Little to none'
        interp += [verdict + ' correlation']

insights = {
    'Correlation coefficient' : coeff,
    'Interpretation' : interp
}

# Bikin Dataframe
result = pd.DataFrame(insights,
                      index = aspect)
# Menunjukkan Hasil
result
```

Kesimpulan
 - Quality Time dan Word of Affirmation berkorelasi sedang
 - Physical Touch and Word of Affirmation berkorelasi sedang
